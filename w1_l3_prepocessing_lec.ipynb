### Dealing with text data like category
<h1> One hot encoding </h1>


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Create a sample DataFrame with multiple columns
data = {
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red'],
    'Size': ['Small', 'Large', 'Medium', 'Medium', 'Small'],
    'Shape': ['Circle', 'Rectangle', 'Triangle', 'Circle', 'Triangle']
}

df = pd.DataFrame(data)

# Initialize the OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform the encoder to the 'Color' column
encoded_data = encoder.fit_transform(df[['Color']])
print(encoded_data.toarray() , encoder.get_feature_names_out(['Color']))
# Convert the sparse matrix to a dense DataFrame
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['Color']))

# Merge the one-hot encoded DataFrame with the original DataFrame
merged_df = pd.concat([df, encoded_df], axis=1)

# Display the merged DataFrame
print("Original DataFrame:")
print(df)
print("\nMerged DataFrame with One-Hot Encoding:")
print(merged_df)
```

##Transformation pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
transform_pipeline=Pipeline(
    [
      ('imputer' , SimpleImputer(strategy='median')),
      ('std_scalar' , StandardScaler())])
```

```python
tr_wine_features = transform_pipeline.fit_transform(wine_features)
```

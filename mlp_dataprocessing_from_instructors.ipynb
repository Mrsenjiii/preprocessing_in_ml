```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

Feature Transformation : It is the process of tranforming the feature into the same scale, as it can help some machine learning model to increase speed.
Some machine learning algorithms like svm,logistic,knn uses eucledian distance so having features in different scale can lead to biasness to some feature by which we can have poor performace as well.

Tree bases algorithms are unafected of scaling, they work on if and else which do no affected by scaling.

Examples of Feature Transformations.



1.   Standardization
2.   Normalization.



```python
x1 = np.random.randint(1, 20, size=(20,1))
x2 = np.random.randint(10,200,size=(20,1))
x3 = np.random.randint(100, 2000, size=(20,1))
X = np.concatenate( (x1,x2,x3 ) , axis = 1) # columns wise addition
```

```python
plt.figure(figsize=(15,8))

# plot the three features
plt.plot(x1)
plt.plot(x2)
plt.plot(x3)

# name the curves of features
plt.legend(['x1','x2','x3'], fontsize=18)


plt.show()
```

```python
X.mean(axis = 0 )
```

```python
X.std(axis = 0)
```

#One Hot Encoding.

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore') # will ignore the other new term when you try to use fitted one to new set.
X = [['Male', 1], ['Female', 3], ['Female', 2]]
enc.fit(X)
enc.categories_
enc.transform([['Female', 1], ['Male', 4]]).toarray()
# enc.inverse_transform([[0, 1, 1, 0, 0], [0, 0, 0, 1, 0]])
enc.get_feature_names_out(['gender', 'group'])
```

```python
drop_enc=OneHotEncoder(drop='first' , sparse_output=False).fit(X)
drop_enc.categories_
drop_enc.transform(X)
```

```python
drop_binary_enc = OneHotEncoder(drop='if_binary' , sparse_output=False).fit(X) # droping only when binary
drop_binary_enc.transform([['Female', 1], ['Male', 2]])
# drop_binary_enc.get_feature_names_out(['gender' ,'group'])
```

#Handeling the Imbalance Dataset.

```python
from sklearn.datasets import make_classification
```

```python
from collections import Counter
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=100)
print(X.shape,y.shape)
print(y)
```

```python
# import  matplotlib.pyplot as plt
# import numpy as np
# plt.figure(figsize=(15,8))
# colors = np.where(y==1,'k','g')
# plt.scatter(x[:,0] , x[:,1] , c=colors, alpha=0.5)
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


colors = np.where(y==1,'k','g')
print(colors)
plt.figure(figsize=(15,8))
plt.scatter(X[:,0], X[:,1], c= colors, alpha=0.6)
plt.show()
```

#UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
Undersample = RandomUnderSampler(sampling_strategy='majority')
x2 , y2 = Undersample.fit_resample(X,y)
```

```python
Counter(y2)
```

```python
plt.figure(figsize=(15,8))
colors=np.where(y2==1,'k','g')
plt.scatter(x2[:,0], x2[:,1], c= colors, alpha=0.6)
plt.show()
```

#OverSampling


```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
```

```python
x1,x2 = oversample.fit_resample(X,y)
```
